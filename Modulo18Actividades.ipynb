```python
def Func1(y,x):
    df1=data_students[x==y]
    gender_scname=df1[["grade","school_name"]]
    dummies_genero=pd.get_dummies(gender_scname[["grade"]])
    plotbygen=pd.concat([gender_scname,dummies_genero], axis=1)
    plot_grouped=plotbygen.groupby(["school_name"]).sum()
    print("Gáfico con puntaje de ", x)
    return plot_grouped.plot(kind="bar", color=["blue","springgreen","Orange",'chocolate'], figsize=(10,4))
```
