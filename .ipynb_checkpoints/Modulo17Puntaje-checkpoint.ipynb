```python
import pandas as pd
import numpy as np
import csv
```

```python
alumnos = pd.read_csv("C:\\Users\\user\\\Documents\\GitHub\\TareasRoel\\cleanstudentscomplete.csv")
```

```python
data_students = alumnos
```

```python
def Func1(y,x):
    df1=data_students[x==y]
    gender_scname=df1[["gender","school_name"]]
    dummies_genero=pd.get_dummies(gender_scname[["gender"]])
    plotbygen=pd.concat([gender_scname,dummies_genero], axis=1)
    plot_grouped=plotbygen.groupby(["school_name"]).sum()
    print("Gáfico con puntaje de ", x)
    return plot_grouped.plot(kind="bar", color=["pink","blue"], figsize=(10,4))
```

```python
y=data_students["reading_score"]
x=data_students["reading_score"].max()
```

```python
Func1(y,x)
```

```python
y=data_students["reading_score"]
x=data_students["reading_score"].min()
```

```python
Func1(y,x)
```
